```python
df = pd.read_csv('train.csv')
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('.','')
df.drop(['ID','Doors','Color'],axis=1,inplace=True)
df.Mileage=df.Mileage.str.replace('km','')
df['Mileage'] = df.Mileage.astype(int)
df['Levy'] = df.Levy.str.replace('-','87')
df.Manufacturer=df.Manufacturer.str.lower()
df['Age'] = 1/(2021 - df.Prod_year)
df.Model=df.Model.str.lower()
df.drop('Prod_year',axis=1,inplace=True)
df['Car_make'] = df.Manufacturer + " "+df.Model
df.drop('Model',axis=1,inplace=True)
print('Mean Price:')

print(df.Price.mean())
df.head()
```

```python
kf = KFold(n_splits  = 10)
print (kf)
train = []
test=[]

for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    train.append(train_index)  
    test.append(test_index) 
    
print(train[0])
```

```python
rf2 = RandomForestRegressor(max_features=10,random_state=10,n_jobs=-1,ccp_alpha=0.25)
yrftr2 = []
yrftt2 = []
rflist = []
for i in range(0,10):
    rflist.append(rf2.fit(X.iloc[train[i]],y.iloc[train[i]]))
    yrftt2.append(rf2.predict(X.iloc[test[i]]))
    print('Test Score',i,' ',np.sqrt(mean_squared_log_error(yrftt2[i],y.iloc[test[i]])))
    yrftr2.append(rf2.predict(X))
    print('Train Score',i,' ',np.sqrt(mean_squared_log_error(yrftr2[i],y)))
yrftr2f = sum(yrftr2)/10

print(np.sqrt(mean_squared_log_error(yrftr2f,y)))
```

```python
etr2 = ExtraTreesRegressor(max_features=10,random_state=10,n_jobs=-1,ccp_alpha=0.25)
yetr2 = []
yett2 = []
etrlist = []
test_scores= []
for i in range(0,10):
    etrlist.append(etr2.fit(X.iloc[train[i]],y.iloc[train[i]]))
    yett2.append(etr2.predict(X.iloc[test[i]]))
    test_scores.append(np.sqrt(mean_squared_log_error(yett2[i],y.iloc[test[i]])))
    print('Test Score',i,' ',test_scores[i])
    yetr2.append(etr2.predict(X))
    print('Train Score',i,' ',np.sqrt(mean_squared_log_error(yetr2[i],y)))
yetr2f = sum(yetr2)/10
yett2f = sum(test_scores)/10
print(np.sqrt(mean_squared_log_error(yetr2f,y)))
print(yett2f)
```
